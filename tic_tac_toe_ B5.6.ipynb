```python
import numpy as np 


n=2   #We can choose the size of the field

field=np.zeros((n, n)) 

check=[]


def who():         #Player generator
    while True:
        yield 'X'
        yield 'O'


def field_display(matrix):     

    print()
    
    for i in range(n):
        print('  ',i, end='')
    
    print()

    for i in range(n):
        print()
        print(i, end='  ')
        for j in range(n):
            if matrix[i][j] == 0: print('-',end='   ')
            elif matrix[i][j] == 1: print('O',end='   ')
            else: print('X',end='   ')
        print()
                

def move (who):
    print()
    print(who, 'move ->')
    
    try:

        row = int(input("Number of row: "))
        column = int(input("Number of column: "))
        
    except Exception as ex: 
        
        print('\nOnly digitals accepted!')
        return False
    
    if  (row not in range (0,n)) or (column not in range (0,n)) or (field[row][column] != 0): 
        
        print ('\nPlease, choose another cell!')
        return False
    
    else:
        
        if  who == 'X': 
            field[row][column]= -1

        else: 
            field[row][column]= 1
            
        field_display(field)
    
        return True
    
    
def victory_check (list):
    
    if n in check:
        print('\n  O is WINNER!!!')
        return True
    
    if -n in check:
        print('\n  X is WINNER!!!')
        return True
                
"""______________________________"""

        
field_display(field)

for x_o in who():   
    
    while True:
        if move(x_o): break
    
    check=[]
    check=list(np.asarray(field).sum(axis=0)) + list(np.asarray(field).sum(axis=1)) #list for check columns and rows
    check.append(sum(field[i][n-i-1] for i in range(n))) #adding check for first diagonal
    check.append(sum(field[i][i] for i in range(n))) #adding check for second diagonal
  

    if victory_check (field): break   #Winner detection
 
    if 0 not in field:      #There is no winner
        print ('\nDraw!')  
        break

       
```
